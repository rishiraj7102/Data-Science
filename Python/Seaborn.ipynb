# Seaborn

## DISTRIBUTION PLOTS

Jointplot<br>
Pairplot<br>
Distplot

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df
```

```python
df.dtypes
```

# Correlation Analysis

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

## Jointplot

```python
sns.jointplot(df['total_bill'],df['tip'],kind='hex')
```

```python
sns.jointplot(df['total_bill'],df['tip'],kind='reg')
```

```python
sns.jointplot(df['total_bill'],df['tip'],kind='kde')
```

## Pairplot


```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="sex")
```

## Distplot

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

# Categorical Plots

Boxplot

Violinplot

Countplot

Barplot

## Countplot

```python
sns.countplot(df.sex)
```

```python
sns.countplot(df.smoker)
```

```python
sns.countplot(df.day)
```

## Barplot

```python
sns.barplot(df.total_bill,df.sex)
```

## Boxplot

```python
sns.boxplot(df.sex,df.total_bill)
```

```python
sns.boxplot(df.day,df.total_bill)
```

```python
sns.boxplot(df.smoker,df.total_bill)
```

```python
sns.boxplot(df.total_bill,df.day,hue=df.smoker)
```

```python
sns.violinplot(df.total_bill,df.sex,hue=df.smoker,color='black')
```
