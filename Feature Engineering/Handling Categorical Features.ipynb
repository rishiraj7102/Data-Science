# Target Guided Ordinal Encoding

```python
df=pd.read_csv("Datasets/titanic/titanic_train.csv",usecols=["Survived","Cabin"])
```

```python
df.head()
```

```python
df.Cabin.fillna("Missing",inplace=True)
```

```python
df.head()
```

```python
df.Cabin=df.Cabin.astype(str).str[0]
```

```python
df.head()
```

```python
df.groupby('Cabin')['Survived'].mean().sort_values()
```

```python
ordinal_labels=df.groupby('Cabin')['Survived'].mean().sort_values().index
```

```python
ordinal_labels
```

```python
ordinal_labels2={k:i for i,k in enumerate(ordinal_labels,0)}
```

```python
ordinal_labels2
```

```python
df["Cabin_ordinal"]=df.Cabin.map(ordinal_labels2)
```

```python
df.head()
```

# mean Encoding

```python
mean_labels=df.groupby('Cabin')['Survived'].mean().sort_values().to_dict()
```

```python
df.Cabin=df.Cabin.map(mean_labels)
```

```python
df.head()
```

# Probability Ratio Encoding  

```python
df=pd.read_csv("Datasets/titanic/titanic_train.csv",usecols=["Survived","Cabin"])
```

```python
df.head()
```

```python
df.Cabin.fillna("Missing",inplace=True)
```

```python
df.Cabin=df.Cabin.astype(str).str[0]
```

```python
df.head()
```

```python
prob_df=pd.DataFrame(df.groupby('Cabin')['Survived'].mean())
```

```python
prob_df
```

```python
prob_df["Died"]=1-prob_df.Survived
```

```python
prob_df
```

```python
prob_df['Ratio']=prob_df.Survived/prob_df.Died
```

```python
prob_df.head()
```

```python
prob_encoded=prob_df.Ratio.to_dict()
```

```python
prob_encoded
```

```python
df['Cabin_encoded']=df.Cabin.map(prob_encoded)
```
