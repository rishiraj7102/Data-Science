# Count/Freq Encoding

```python
df=pd.read_csv('Datasets/mercedes/train.csv',usecols=['X1','X2'])
```

```python
df
```

```python
df_freq_map=df.X2.value_counts().to_dict()
```

```python
df.X2=df.X2.map(df_freq_map)
```
