```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
    df=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",names=['sepal_length','sepal_width','petal_length','petal_width','species'])
```

```python
df.head()
```

```python
array=[]
for i in df.species:
    array.append(i.split('-')[1])
```

```python
array
```

```python
df.species=array
```

```python
df.head()
```

# univariate

```python
df.species.unique()
```

```python
df_setosa=df.loc[df.species=='setosa']
df_versicolor=df.loc[df.species=='versicolor']
df_virginica=df.loc[df.species=='virginica']
```

```python
df_setosa.head()
```

```python
plt.plot(df_sentosa['sepal_length'],np.zeros_like(df_sentosa['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'o')
```

# bivariate 

```python
sns.FacetGrid(df,hue='species',size=5).map(plt.scatter,'petal_length','sepal_width').add_legend()
```
